# Results of fifth trial

# No Aggregation, 2 robots

```python
extension = 'csv'
all_filenames = [i for i in glob.glob('./5th_trial/e50_fp30/2r/nosl/*.{}'.format(extension))]

columns=['discovery','attainment','total_time','Location',]
cdf = pd.DataFrame(columns=columns)

i=0
dfs = []
for f in all_filenames:
    if 'combined_data' in f:
        continue
    df = pd.read_csv(f)
    dfs.append(pd.DataFrame(df, columns=['discovery','attainment','total_time']).assign(Location='Run'+str(i+1)))
    i = i+1

cdf = pd.concat(dfs)
mdf = pd.melt(cdf, id_vars=['Location'], var_name=['Metric'])

ax = sns.boxplot(x="Location", y="value", hue='Metric', data=mdf, notch=False, width=0.65)
#sns.swarmplot(x="Location", y="value", data=mdf, color=".25")

#ax = ax.axes
#lines = ax.get_lines()
#categories = ax.get_xticks()

#for cat in categories:
#    # every 4th line at the interval of 6 is median line
#    # 0 -> p25 1 -> p75 2 -> lower whisker 3 -> upper whisker 4 -> p50 5 -> upper extreme value
#    y = round(lines[4+cat*6].get_ydata()[0],1) 

#    ax.text(cat, y, f'{y}', ha='center', va='center', fontweight='bold', size='x-small', color='black')

#plt.show()
plt.xlabel('No Aggregation (2 robots)')
plt.ylabel('Time in s')
plt.savefig('./5th_trial/e50_fp30/2r/nosl/bplot_2r_5run.png', bbox_inches='tight', dpi=300)
```
